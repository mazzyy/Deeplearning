```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create checkers board 
# Create a 4x4 matrix of zeros
matrix = np.zeros((10, 10), dtype=int)

```

```python
matrix[::2, ::2] = 1
matrix[1::2, 1::2] = 1
```

```python
matrix
```

```python
# 2d dimension heatmap (CheakBoard)
plt.imshow(matrix, cmap='hot')
plt.show()
```

```python
# (0,2) are center asix of circle where 1 is radious 
circle = plt.Circle ((0, 2), 1,color="k")

fig, ax = plt.subplots()

ax.axis ("equal")
ax.set_facecolor('black')

# Set the plot limits of x and y axis
ax.set_xlim ([-5,5]) 
ax.set_ylim([-5,5])  
ax.add_patch(circle) # add path circle
```
